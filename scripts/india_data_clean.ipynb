```python
dfz.columns = dfz.columns.str.lower().str.replace(" ", "_")
```

```python
dfz.columns
```

```python
dfz.shape
```

```python
df = dfz.groupby(
    ['restaurant_name', 'dining_rating', 'cuisine_', 'place_name', 'city']
).agg({
    "item_name": lambda x: ", ".join(sorted(x.unique())),
    "prices": "median"
}).reset_index()
```

```python
df.head()
```

```python
df.tail()
```

```python
df.columns
```

```python
df.rename(columns={
    'dining_rating': 'rating',
    'cuisine_': 'cuisines',
    'item_name': 'menu_list'
}, inplace=True)
```

```python
df['restaurant_name'] = df['restaurant_name'].astype(str).str.strip()
df['city'] = df['city'].astype(str).str.strip()
df['place_name'] = df['place_name'].astype(str).str.strip()
```

```python
# SAVE CLEAN FILE
df.to_csv("raw_data/zomato_india.csv", index=False)

print("CLEANED SHAPE:", df.shape)
```
